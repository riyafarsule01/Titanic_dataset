```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# --- Load the Dataset ---
df = pd.read_csv(r"C:\Users\riyas\Downloads\archive\Titanic-Dataset.csv")
```

```python
# --- Step 1: Initial Data Inspection (.info(), .head()) ---
print("--- Step 1: Initial Data Inspection ---")
print("First 5 Rows of the Dataset:")
print(df.head())
print("\nDataset Information (Non-Null Counts and Dtypes):")
df.info()
```

```python
# --- Step 4: Visualizations for Insights ---
print("\n--- Generating Visualizations ---")

# Visualization 1: Survival Count
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df, palette='pastel')
plt.title('Distribution of Survival (0 = No, 1 = Yes)', fontsize=16)
plt.xticks([0, 1], ['Did not Survive', 'Survived'])
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.savefig('survival_distribution.png')
plt.show()
```

```python
# Visualization 2: Survival Rate by Sex
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df, palette='muted')
plt.title('Survival Rate by Sex', fontsize=16)
plt.ylabel('Survival Rate')
plt.xlabel('Sex')
plt.savefig('survival_by_sex.png')
plt.show()
```

```python
# Visualization 3: Survival Rate by Passenger Class
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df, palette='deep')
plt.title('Survival Rate by Passenger Class', fontsize=16)
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.savefig('survival_by_pclass.png')
plt.show()
```

```python
# Visualization 4: Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), kde=True, bins=30, color='skyblue')
plt.title('Distribution of Passenger Ages', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('age_distribution.png')
plt.show()
```

```python
# Visualization 5: Correlation Heatmap
plt.figure(figsize=(10, 7))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features', fontsize=16)
plt.savefig('correlation_heatmap.png')
plt.show()
```

```python
# ---  Pairplot Visualization ---
# Using a subset of columns for a clearer plot
pairplot_df = df[['Survived', 'Pclass', 'Age', 'Fare', 'Sex']].dropna()
sns.pairplot(pairplot_df, hue='Survived', palette='viridis', diag_kind='kde')
plt.suptitle('Pairplot of Key Titanic Features by Survival Status', y=1.02, fontsize=16)
plt.show()
```

```python
# --- Boxplot ---
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=df, palette='pastel')
plt.title('Boxplot of Age by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# ---  Scatterplot ---
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='bright')
plt.title('Scatterplot of Age vs. Fare', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived (0=No, 1=Yes)')
plt.show()
```
